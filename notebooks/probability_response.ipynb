```python
def presp_1(sigma=0, sigma_1=3, sigma_2=15, x=None):
    p1 = norm.pdf(x, loc=0, scale=sigma_1+sigma)
    p2 = norm.pdf(x, loc=0, scale=sigma_2+sigma)
    p1 = p1 / np.sum(p1)
    p2 = p2 / np.sum(p2)
    return np.sum((p1 > p2) * 0.5 * (p1 + p2))
```

```python
x = np.linspace(-100, 100, 100000)
```

```python
sigmas = np.linspace(0, 10, 100)
```

```python
pr = [presp_1(sigma, x=x) for sigma in sigmas]
```

```python
plt.plot(sigmas, pr)
```
